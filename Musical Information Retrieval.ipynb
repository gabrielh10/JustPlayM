Retirando Nome e Ids de todas as músicas da playlist e guardando em arrays

```python
AllMusicsIds = []
AllMusicNames = []

def extractSongsName(playlist):
    musicNames = []
    for track in playlist ['tracks']['items']:
        musicNames.append(track['track']['name'])
    return musicNames
def extractSongsID(playlist):    
    musicIds = []
    for track in playlist ['tracks']['items']:
        musicIds.append(track['track']['id'])
    return musicIds
#musicIds 
targetNames = extractSongsName(targetPlaylist)
targetIds = extractSongsID(targetPlaylist)

#musicNames = extractSongsName(playlists)
#musicIds = extractSongsID(playlists)

targetIds
```

Acessando as features de audio das músicas, cujos nomes e ids foram colhidos anteriormente

```python
dataMusicFeatures = pd.DataFrame()
musicFeatures = []
for i in range(0, len(targetIds),1):
    musicFeature = spotify.audio_features(targetIds[i])
    musicFeatures.append(musicFeature)
    
   # dfItem = jsonToDataFrame(data)                
    dataMusicFeatures = dataMusicFeatures.append(musicFeature, ignore_index=True) 
  #  dataFrameHistorical = dataFrameHistorical.append(dfItem, ignore_index=True)
dataMusicFeatures
#musicFeatures
```

```python
dataMusicFeaturesToLearn = dataMusicFeatures.drop(columns=['analysis_url', 'id', 'track_href', 'type', 'uri'])
dataMusicFeaturesToLearn
```

```python
numberOfClusters = widgets.IntText(description="Nº Clusters", continuous_update=False) 
numberOfClusters
```

```python
def runKmeans(nClusters):
    model = KMeans(n_clusters = nClusters, random_state = 7)
    model.fit(dataMusicFeaturesToLearn)
    clusters = model.labels_
    dataMusicFeatures['Cluster'] = clusters
    dataMusicFeatures['Nome'] = targetNames
    dataMusicFeatures
    
runKmeans(numberOfClusters.value)
dataMusicFeatures
```
