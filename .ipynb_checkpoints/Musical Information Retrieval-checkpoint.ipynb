Inicializando array de músicas e preparando para extrair músicas das playlists desejadas


```python
def extractFromPlaylist(user, idPlaylist):
    playlist = spotify.user_playlist(user, idPlaylist)
    return playlist

exalta = extractFromPlaylist("luizapss", '5CTHNQtDeDVCVbBGQjpPfP')



```

Retirando Nome e Ids de todas as músicas da playlist e guardando em arrays

```python
def extractSongsName(playlist):
    musicNames = []
    for track in playlist ['tracks']['items']:
        musicNames.append(track['track']['name'])
    return musicNames
def extractSongsID(playlist):    
    musicIds = []
    for track in playlist ['tracks']['items']:
        musicIds.append(track['track']['id'])
    return musicIds
#musicIds 
musicNames = extractSongsName(exalta)
musicIds = extractSongsID(exalta)
musicIds
```

Acessando as features de audio das músicas, cujos nomes e ids foram colhidos anteriormente

```python
dataMusicFeatures = pd.DataFrame()
musicFeatures = []
for i in range(0, len(musicIds),1):
    musicFeature = spotify.audio_features(musicIds[i])
    musicFeatures.append(musicFeature)
    
   # dfItem = jsonToDataFrame(data)                
    dataMusicFeatures = dataMusicFeatures.append(musicFeature, ignore_index=True) 
  #  dataFrameHistorical = dataFrameHistorical.append(dfItem, ignore_index=True)
dataMusicFeatures
#musicFeatures
```

Limpando Atributos que não vão ser usados na aprendizagem

```python
dataMusicFeaturesToLearn = dataMusicFeatures.drop(columns=['analysis_url', 'id', 'track_href', 'type', 'uri'])
dataMusicFeaturesToLearn
```

Testando kmeans pra clusterizar em românticas ou agitadas

```python
model = KMeans(n_clusters = 2, random_state = 7)
model.fit(dataMusicFeaturesToLearn)
clusters = model.labels_
dataMusicFeatures['Cluster'] = clusters
dataMusicFeatures['Nome'] = musicNames
dataMusicFeatures
```
